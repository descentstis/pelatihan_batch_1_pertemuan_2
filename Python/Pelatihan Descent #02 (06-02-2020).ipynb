# crosstab

```python
survived_sex = pd.crosstab(index=data["survived"], columns=[data["sex"], data['embarked']])
survived_sex
```

```python
survived_class = pd.crosstab(index=data["survived"], columns=data["pclass"])
survived_class.columns = ["class1","class2","class3"]

survived_class
```

```python
# Table of survival vs passenger class
survived_class = pd.crosstab(index=data["survived"], columns=data["pclass"], margins=True)   # Include row and column totals

survived_class.columns = ["class1","class2","class3","rowtotal"]
survived_class.index= ["died","survived","coltotal"]

survived_class
```

```python
survived_class/survived_class.loc["coltotal","rowtotal"]
```

```python
survived_class.mean(axis = 0)
```

```python
survived_class.mean(axis = 1)
```

# Combining Data

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],

                    'value': [1, 2, 3, 5]})

df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],

                    'value': [5, 6, 7, 8]})
```

```python
df1
```

```python
df2
```

## Merge

```python
df1.merge(df2, left_on='lkey', right_on='rkey')
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey',suffixes=('_left', '_right'))
```

## Join

```python
df1.join(df2, lsuffix='_caller', rsuffix='_other')
```

# Pandas Apply & Assign (Mutate in dplyr R)

```python
df = pd.DataFrame({
    'name': ['alice','bob','charlie','daniel'],
    'age': [25,66,56,78]})
```

```python
df
```

## Apply

```python
df['birth_year'] = df['age'].apply(lambda x: 2020-x)
df
```

## Assign

```python
df.assign(is_senior = lambda dataframe: dataframe['age'].map(lambda age: 'Senior' if age >= 65 else False))
```
